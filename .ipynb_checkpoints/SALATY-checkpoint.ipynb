```python
from bs4 import BeautifulSoup as bs
import itertools
from itertools import combinations
import re
import pandas as pd
import numpy as np
import copy
import matplotlib.pyplot as plt
```

```python
file = open('salaty.csv')
salaty = pd.read_csv(file, index_col='name')
```

```python
salaty
```

```python
df_asint = salaty.astype(int)
coocc = df_asint.T.dot(df_asint)
```

```python
np.fill_diagonal(coocc.values, 0)
```

```python
coocc
```

```python
def table_2_3column(table, to_dataframe=True):
    new_table = []
    names = table.columns
    for i in range(len(table)):
        for j in range(i, len(table)):
            if table.iloc[i, j] != 0:
                new_table.append([names[i], names[j], table.iloc[i, j]])
    if to_dataframe:
        return pd.DataFrame(new_table, columns=('Source', 'Target', 'Weight'))
    return new_table
```

```python
table = table_2_3column(coocc)
```

```python
table
```

```python
table.to_csv('Table.csv', index = False, sep='\t')
```
